```python
class CicularQueue:
  def __init__(self, size):
    self.size = size
    self.q = [None] * size
    self.front = 0
    self.rear = 0

  def enqueue(self, data):
    if (self.rear + 1) % self.size == self.front:
        print("Queue is full, can't enqueue")
    else:
        self.q[self.rear] = data
        self.rear = (self.rear + 1) % self.size
        print(f"{data} added to the queue")

  def dequeue(self):
    if self.front == self.rear:
        print("Queue is empty, can't dequeue")
        return None

    dequeued_data = self.q[self.front]
    self.front = (self.front + 1) % self.size
    print(f"{dequeued_data} has been deleted from the queue")
    return dequeued_data

  def is_empty(self):
    return self.front == self.rear

  def size_of_queue(self):
    return (self.rear - self.front + self.size) % self.size

  def peek(self):
    if self.is_empty():
        print("Queue is empty")
        return None
    return self.q[self.front]

  def display(self):
    if self.is_empty():
        print("Queue is empty")
    else:
        elements = []
        i = self.front
        while i != self.rear:
            elements.append(self.q[i])
            i = (i + 1) % self.size
        print(f"Queue: {elements}")
```

```python
Q = CicularQueue(5)
Q.enqueue(5)
Q.enqueue(6)
Q.enqueue(7)
Q.enqueue(8)
```

```python
Q.display()
```
