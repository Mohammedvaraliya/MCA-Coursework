```python
class Queue:
  def __init__(self):
    self.q = []
    self.front = 0
    self.rear = 0

  def enqueue(self, data):
    self.q.append(data)
    self.rear += 1
    print(f"{data} added to the queue")

  def dequeue(self):
    if self.front == self.rear:
        print("Queue is empty, can't dequeue")
        return None

    dequeued_data = self.q[self.front]
    self.front += 1
    print(f"{dequeued_data} has been deleted from the queue")

    if self.front == self.rear:
        self.front = self.rear = 0

  def is_empty(self):
    return self.front == self.rear

  def size(self):
    return self.rear - self.front

  def peek(self):
    if self.is_empty():
        print("Queue is empty")
        return None
    return self.q[self.front]

  def display(self):
    if self.is_empty():
        print("Queue is empty")
    else:
        print(f"Queue: {self.q[self.front:self.rear]}")
```

```python
Q = Queue()
Q.enqueue(5)
Q.enqueue(6)
Q.enqueue(7)
Q.enqueue(8)
```

```python
Q.display()
```

```python
Q.dequeue()
```

```python
Q.display()
```
