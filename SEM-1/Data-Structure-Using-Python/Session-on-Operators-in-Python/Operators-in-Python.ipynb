#### Aim:

Define and understand the different types of operators in Python, providing code examples and outputs for each.

#### Concept

**Operators** are special symbols that perform operations on variables and values. In Python, operators are categorized by the type of operation they perform. Here are the primary types:

1. **Arithmetic Operators**: for basic mathematical operations.
2. **Comparison Operators**: to compare values.
3. **Logical Operators**: to combine conditional statements.
4. **Assignment Operators**: to assign values to variables.
5. **Bitwise Operators**: for operations on binary numbers.
6. **Identity Operators**: to check if two objects are the same.
7. **Membership Operators**: to check for membership in sequences like lists or strings.

#### Arithmetic Operators

```python
a = 10
b = 3

# Addition
print("a + b =", a + b)

# Subtraction
print("a - b =", a - b)

# Multiplication
print("a * b =", a * b)

# Division
print("a / b =", a / b)

# Modulus (remainder)
print("a % b =", a % b)

# Exponentiation
print("a ** b =", a ** b)

# Floor Division
print("a // b =", a // b)
```

#### Comparison Operators

```python
a = 10
b = 3

print("a == b:", a == b)   # Equal to
print("a != b:", a != b)   # Not equal to
print("a > b:", a > b)     # Greater than
print("a < b:", a < b)     # Less than
print("a >= b:", a >= b)   # Greater than or equal to
print("a <= b:", a <= b)   # Less than or equal to
```

#### Logical Operators

```python
a = True
b = False

print("a and b:", a and b)  # Logical AND
print("a or b:", a or b)    # Logical OR
print("not a:", not a)      # Logical NOT
```

#### Assignment Operators

```python
x = 5       # Assign

x += 3      # Add and assign
print("x after x += 3:", x)  # Output: 8

x -= 2      # Subtract and assign
print("x after x -= 2:", x)  # Output: 6

x *= 4      # Multiply and assign
print("x after x *= 4:", x)  # Output: 24

x /= 2      # Divide and assign
print("x after x /= 2:", x)  # Output: 12.0

x %= 3      # Modulus and assign
print("x after x %= 3:", x)  # Output: 0.0

x **= 2     # Exponent and assign
print("x after x **= 2:", x) # Output: 0.0

x //= 3     # Floor division and assign
print("x after x //= 3:", x) # Output: 0.0
```

#### Bitwise Operators

```python
a = 5  # (binary: 0101)
b = 7  # (binary: 0011)

print("a & b:", a & b)   # AND
print("a | b:", a | b)   # OR
print("a ^ b:", a ^ b)   # XOR
print("~a:", ~a)         # NOT
print("a << 1:", a << 1) # Left Shift
print("a >> 1:", a >> 1) # Right Shift
```

#### Identity Operators

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print("a is b:", a is b)         # True (same object)
print("a is not c:", a is not c) # True (different objects)
```

#### Membership Operators

```python
my_list = [1, 2, 3, 4]
print(2 in my_list)     # True
print(5 not in my_list) # True
```
