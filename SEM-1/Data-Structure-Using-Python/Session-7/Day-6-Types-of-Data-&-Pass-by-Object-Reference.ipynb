## Immutable: The data of original variable won't change after assignment

```python
a = 1
b = a + 2
```

```python
a
```

```python
b
```

## Mutable: The data will change if you add or delete elements from original list after assignment

```python
a = [1, 2, 3]
b = a
```

```python
a.append(4)
```

```python
a
```

```python
b # The data is automatically updated
```

## Pass by object reference

```python
def swap(nums):
  temp = nums[0]
  nums[0] = nums[1]
  nums[1] = temp
```

```python
numbers = [10, 20]
print("Before swapping Numbers: ", numbers)
```

```python
swap(nums=numbers)
print("After swapping Numbers: ", numbers)
```

## Recursive Functions

```python
def fact(n):
  if(n == 0):
    return 1
  else:
    return n * fact(n-1)
```

```python
n = int(input("Enter a number for finding factorial: "))
print(f"The factorial of number {n} is: ", fact(n))
```
