### Aim:

To implement **Linear Search** in Python to find the position of a target element within a list by sequentially checking each element.

### Concept:

Linear Search is a simple searching algorithm used to find an element in a list. It sequentially checks each element of the list until the target element is found or the end of the list is reached. Linear search is efficient for small lists but can be slow for large lists due to its time complexity of $O(n)$, where $n$ is the number of elements in the list.

### Algorithm:

In the algorithm, we iterate through each element of the list and compare it with the target element. If a match is found, the search stops, and the position of the target element is returned. If the loop completes without finding the target, it indicates that the element is not in the list.

1. **Step 1**: Start from the first element of the list.
2. **Step 2**: Check each element sequentially:
    1. If the current element matches the target element, return its index (position).
    2. If it doesn't match, move to the next element.
3. **Step 3**: If the target element is not found by the end of the list, return -1 to indicate it is absent.

#### Hereâ€™s the Python implementation for Linear Search. This code searches for a target element within a given list and returns the index of the element if found; otherwise, it returns -1.

```python
def linear_search(arr, target):
    for index in range(len(arr)):
        if arr[index] == target:
            return index
    return -1

# Test the function
arr = [4, 2, 7, 1, 9, 3]
target = 7
result = linear_search(arr, target)

# Display the result
if result != -1:
    print(f"Element {target} found at index: {result}")
else:
    print(f"Element {target} not found in the list")
```
