```python
def push_ele(stk, ele):
  global top
  top += 1
  stk.insert(top, ele)
  return stk, top
```

```python
def pop_ele():
  global top
  if top == -1:
    print("Stack underflow")
  else:
    ele = stk.pop()
    top -= 1
    return ele
```

```python
stk = [10, 20, 30, 40, 50]
top = len(stk) - 1

push_ele(stk, 60)
print(stk)
```

```python
print("Popped element", pop_ele())
print(stk)
```
