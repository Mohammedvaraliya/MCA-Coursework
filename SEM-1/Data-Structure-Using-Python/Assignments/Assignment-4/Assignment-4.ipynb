### Question-1. Write a program to accept a number from the user and display whether it is even or odd using function.

```python
def check_num(num):
  if(num%2==0):
    print("Even")
  else:
    print("Odd")
```

```python
check_num(10)
```

### Question-2: Create a function to display the given number is of one digited or two digited or three digited or more than three digited.

```python
def check_digit(num):
  if(num<10):
    print("One digited")
  elif(num<100):
    print("Two digited")
  elif(num<1000):
    print("Three digited")
  else:
    print("More than three digited")
```

```python
check_digit(1000)
```
