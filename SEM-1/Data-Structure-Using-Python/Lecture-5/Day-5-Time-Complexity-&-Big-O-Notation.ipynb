### Question-1: Write a program to accept the distance covered by your car and the time taken to reach your destination, find the speed of your car.

Formula of Speed = dist/time

```python
dist = int(input("Enter Distance: "))
time = int(input("Enter Time: "))

speed = dist/time
print("Speed of the car= ", speed, "Km/Hr")
```

```python
# Now the same above program with function

def speed(dist, time):
    speed = dist/time
    print("Speed of the car= ", speed, "Km/Hr")
```

```python
dis = int(input("Enter Distance: "))
time = int(input("Enter Time: "))

speed(dist, time)
```

The time complexity of speed function is O(1)

### Question-2: Calculate the area of Circle using function.

Formula of area of circle = Pi(r^2)

```python
# Without return & without parameter

def Calc_Area():
  r = int(input("Enter Radius of a circle: "))
  area = 3.14*r*r
  print("Area of a circle = ", area)
```

```python
Calc_Area()
```

```python
# With return

def Calc_Area():
  r = int(input("Enter Radius of a circle: "))
  area = 3.14*r*r
  return area
```

```python
res = Calc_Area()
print("Area of a circle = ", res)
```

```python
# We can club the function call in a print statement too, that'll reduce memory space

print("Area of a circle = ", Calc_Area())
```

```python
# Rewrite the area of circle code with return & with parameter or arguments

def Calc_Area(radius):
  area = 3.14*radius*radius
  return area
```

```python
radius = int(input("Enter Radius of a circle: "))
print("Area of a circle = ", Calc_Area(radius))
```

### Question. Write a program to accept a number from the user and display whether it is even or odd using function.

### Question: Create a function to display the given number is of one digited or two digited or three digited or more than three digited.

```python
def check_digit(num):
  if(num<10):
    print("One digited")
  elif(num<100):
    print("Two digited")
  elif(num<1000):
    print("Three digited")
  else:
    print("More than three digited")
```

```python
check_digit(1000)
```

### Store monthly expenses in a list and find out the total expenses for all months

```python
exp = [12500, 15000, 13500, 14590, 13700]
total = 0

for element in exp:
  total = total + element

print("Total expenses for all months = ", total)
```

```python
# Do the same operation using function to make things easy

def calc_total(exp):
  total = 0
  for element in exp:
    total = total + element
  return total
```

```python
amol_exp = [12500, 15000, 13500, 14590, 13700]
print("Total expenses for Amol = ", calc_total(amol_exp))
```

```python
anand_exp = [13500, 12000, 1500, 13540, 17600]
print("Total expenses for Anand = ", calc_total(anand_exp))
```

### Write a python function to sum all the numbers in a list

```python
# Way 1 (Without return)

def sum(nums):
  total = 0
  for num in nums:
    total += num

  print("The sum of all numbers in a list are: ", total)
```

```python
nums1 = [1, 3, 4, 5, 7, 8]
sum(nums1)
```

```python
# Way 2 (With return)

def sum(nums):
  total = 0
  for num in nums:
    total += num

  return total
```

```python
nums2 = [1, 3, 4, 5, 7, 8, 10, 20, 1000, 10]
res2 = sum(nums2)
print("The sum of all numbers in a list are: ", res2)
```
