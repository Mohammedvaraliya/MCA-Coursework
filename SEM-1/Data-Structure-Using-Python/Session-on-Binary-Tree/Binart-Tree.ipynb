```python
class Node:
  def __init__(self, data):
    self.data = data
    self.lchild = None
    self.rchild = None
```

```python
def level_order_traversal(root):
    def height(node):
        if not node:
            return 0
        return 1 + max(height(node.lchild), height(node.rchild))

    def print_level(node, level):
        if not node:
            return
        if level == 1:
            print(node.data, end=" ")
        elif level > 1:
            print_level(node.lchild, level - 1)
            print_level(node.rchild, level - 1)

    h = height(root)
    for i in range(1, h + 1):
        print_level(root, i)
```

```python
def inorder_traversal(root):
    if root:
        inorder_traversal(root.lchild)
        print(root.data, end=" ")
        inorder_traversal(root.rchild)
```

```python
def preorder_traversal(root):
    if root:
        print(root.data, end=" ")
        preorder_traversal(root.lchild)
        preorder_traversal(root.rchild)
```

```python
def postorder_traversal(root):
    if root:
        postorder_traversal(root.lchild)
        postorder_traversal(root.rchild)
        print(root.data, end=" ")
```

```python
def create():
    print("Enter root value: ")
    x = int(input())
    root = Node(x)

    def insert_children(node):
        if node:
            print(f"Enter left child of {node.data} (enter -1 if no left child): ")
            x = int(input())
            if x != -1:
                node.lchild = Node(x)
                insert_children(node.lchild)

            print(f"Enter right child of {node.data} (enter -1 if no right child): ")
            x = int(input())
            if x != -1:
                node.rchild = Node(x)
                insert_children(node.rchild)

    insert_children(root)
    return root
```

```python
root = create()

print("\nLevel-order Traversal:")
level_order_traversal(root)

print("\nIn-order Traversal:")
inorder_traversal(root)

print("\nPre-order Traversal:")
preorder_traversal(root)

print("\nPost-order Traversal:")
postorder_traversal(root)
```
