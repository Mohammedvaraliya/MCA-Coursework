### **Aim:**

To understand and implement methods for finding the maximum and minimum values in a list using Pythonâ€™s built-in functions and custom code.

### **Concept:**

The concept of finding the maximum or minimum value involves examining each element in a dataset (e.g., list) and identifying the largest or smallest element. Python provides built-in functions like `max()` and `min()` that simplify this process, but we can also write custom functions to achieve the same result.

### **Algorithm:**

1. **Using Built-in Functions (`max()` and `min()`)**
    1. Start with a list of values.
    2. Call the `max()` function to find the maximum value.
    3. Call the `min()` function to find the minimum value.
    4. Display the results.
    
    ---
    
2. **Using a Custom Loop-Based Function**
    1. Start with a list of values.
    2. Initialize two variables to hold the maximum and minimum values, starting with the first element of the list.
    3. Traverse through each element:
        - If an element is greater than the current maximum, update the maximum value.
        - If an element is less than the current minimum, update the minimum value.
    4. Display the final maximum and minimum values.
    
    ---

#### Example 1: Using Built-in max() and min() Functions

```python
# List of numbers
numbers = [3, 7, 2, 8, 5, 10, 1]

# Find maximum and minimum using built-in functions
max_value = max(numbers)
min_value = min(numbers)

# Display the results
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

#### Example 2: Finding Maximum and Minimum Using a Custom Function

```python
# List of numbers
numbers = [3, 7, 2, 8, 5, 10, 1]

# Initialize maximum and minimum values with the first element of the list
max_value = numbers[0]
min_value = numbers[0]

for num in numbers:

    if num > max_value:
        max_value = num

    if num < min_value:
        min_value = num

# Display the results
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```
