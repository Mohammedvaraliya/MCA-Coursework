## Swap of two numbers

```python
def swap(a, b):
  temp = a
  a = b
  b = temp
  print("a = ", a)
  print("b = ", b)
```

```python
m = 10
n = 20
print("m = ", m)
print("n = ", n)
```

```python
swap(m, n)
```
