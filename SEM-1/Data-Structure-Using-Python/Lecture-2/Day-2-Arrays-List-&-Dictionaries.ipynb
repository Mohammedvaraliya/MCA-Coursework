```python
dict = {
    "Asha": 8769777888,
    "Usha": 9819666555,
    "Disha": 7039222444
}
```

```python
dict
```

```python
# Creating Dictionary using list

details = {
    "Name": ['Mohd', "Shaun", "Anup"],
    "Age": [22, 21, 27],
    "University": ["JNU", "DHU", "DU"]
}
```

```python
details
```

```python
details["Name"][0]
```

```python
[elem[0] for elem in details.values()]
```

```python
# Retrieve the first value from each list
first_values = {key: values[0] for key, values in details.items()}

print(first_values)
```
