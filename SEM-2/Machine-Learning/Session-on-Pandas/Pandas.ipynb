```python
import pandas as pd
```

```python
s1 = pd.Series([10, 20, 30], index=["Jan", "Feb", "Mar"])
```

```python
s1
```

### Creating DataFrame using Multiple Series

```python
s2 = pd.Series([100, 200, 300], index=["Jan", "Feb", "Mar"])
df3 = pd.DataFrame([s1, s2])
```

```python
df3
```

```python
df4 = pd.DataFrame([s1, s2], index=['Sale1', 'Sale2'])
```

```python
df4
```

### Create dataframe passing dictionary using key value pair

```python
s3 = pd.Series(['Kabaddi', 'Kho-Kho', 'Volley Ball'], index=['R1', 'R2', 'R3'])
s4 = pd.Series([7, 12, 5], index=['R1', 'R2', 'R3'])
dict1 = {'Sports': s3, 'No of Players': s4}
```

```python
df5 = pd.DataFrame(dict1)
```

```python
df5
```

```python
dic2 = {
    'Manager': {
        'IT Support': 'Raman', 'Operations': 'Kurti'
    },
    'Engineer': {
        'IT Support': 'Jayesh', 'Operations': 'Shreyash'
    }
}
```
