### KNN on Iris Dataset

```python
uploaded2 = files.upload()
filename2 = next(iter(uploaded2))
```

```python
iris_df = pd.read_csv(filename2)
```

```python
iris_df.info()
```

```python
iris_df
```

```python
X = iris_df.drop(['variety'], axis=1)
y = iris_df.variety
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=50)
knn
```

```python
train_df = knn.fit(X_train, y_train)
train_df
```

```python
y_pred = knn.predict(X_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted')}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted')}")
print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted')}")
```

### KNN on Social Network Ads Dataset

```python
uploaded3 = files.upload()
filename3 = next(iter(uploaded3))
```

```python
social_network_df = pd.read_csv(filename3)
```

```python
social_network_df.info()
```

```python
social_network_df
```

```python
X = social_network_df.drop(['Purchased', 'User ID',	'Gender'], axis=1)
y = social_network_df.Purchased
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn
```

```python
train_df = knn.fit(X_train, y_train)
train_df
```

```python
y_pred = knn.predict(X_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted')}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted')}")
print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted')}")
```

```python
print(classification_report(y_test, y_pred))
```

### Hyper Parameter Tuning


A hyperparameter is a parameter of the model that is set before the start of learing process. Different machine learning models have different hyperparameters.

The main purpose of hyperparameter tuning in K-Nearest Neighbors (KNN) is to find the best settings that give your model the highest accuracy and generalization to unseen data.

```python
grid_params = {
    'n_neighbors': [5, 7, 9, 11, 13, 15],
    'metric': ['minkowski', 'euclidean', 'manhattan']
}
```

```python
gs = GridSearchCV(knn, grid_params, verbose=2, cv=3)
```

```python
gs
```

```python
g_res = gs.fit(X_train, y_train)
g_res
```

```python
g_res.best_score_
```

```python
g_res.best_params_
```
