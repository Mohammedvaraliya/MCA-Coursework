### Label Encoding

```python
le = LabelEncoder()
```

```python
pets_df['Animals'] = le.fit_transform(pets_df['Animals'])
pets_df['Size of Animal'] = le.fit_transform(pets_df['Size of Animal'])
pets_df['Body Color'] = le.fit_transform(pets_df['Body Color'])
```

```python
X = pets_df.drop(['Can we pet them'], axis=1)
```

```python
X
```

```python
y
```

```python
berNB = CategoricalNB()
berNB
```

```python
train_df = berNB.fit(X, y)
train_df
```

### Prediction 1

Animals = Rat (2), Size of Animal = Big (0), Body Color = White (2)
