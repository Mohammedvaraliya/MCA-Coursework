## Histogram

```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

### How many patients are normal?

```python
# Normal range (70-99)
normal = [x for x in blood_sugar if 70 <= x <= 99]
len(normal)
```

### How many of them are pre-diabetic?

```python
# Pre-diabetic range (100-125)
pre_diabetic = [x for x in blood_sugar if 100 <= x <= 125]
len(pre_diabetic)
```

### Pre-diabatic (less than 100) and (above 125) i.e diabatic

```python
# Diabetic (above 125)
diabetic = [x for x in blood_sugar if x > 125]
len(diabetic)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.xlabel('Blood Sugar Levels')
plt.ylabel('Number of Patients')
plt.title('Histogram of Blood Sugar Levels')
plt.hist(blood_sugar, bins=[80, 100, 125, 150], edgecolor='black')
plt.show()
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 49, 89, 79, 129, 129, 112, 100]
```

```python
plt.xlabel('Blood Sugar Levels')
plt.ylabel('Number of Patients')
plt.title('Histogram of Blood Sugar Levels')

plt.hist([blood_sugar, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, label=['Men', 'Women'], edgecolor='black', orientation='horizontal')
plt.legend()
```
