**Finding the best fit line means that "we should calculate an equation of a line that best fits the data" where all points are as close as possible to that line.**

In machine learning, a **loss function** measures how well a model's predictions match the actual data. Here's the general formula for a loss function in markdown equation format:

### Mean Squared Error (MSE) Loss:
$$L(\theta) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
- $L(\theta)$ is the loss.
- $n$ is the number of data points.
- $y_i$ is the actual value.
- $\hat{y}_i$ is the predicted value.

### Cross-Entropy Loss (for classification problems):
$$L(\theta) = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$

Where:
- $y_i$ is the true label (0 or 1).
- $\hat{y}_i$ is the predicted probability of the positive class.

These are two common types of loss functions used in regression and classification problems, respectively.
