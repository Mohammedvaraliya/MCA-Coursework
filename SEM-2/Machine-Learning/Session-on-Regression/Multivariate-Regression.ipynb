```python
bedroomMean = houseData.Bedroom.mean()
bedroomMean
```

### Replace NaN with mean of Bedroom data

```python
houseData.Bedroom = houseData.Bedroom.fillna(value=bedroomMean)
houseData
```

### 2. Develop a model using multiple regression.

#### Create an Object for Linear Regression

```python
reg = linear_model.LinearRegression()
```

#### Fit the Data

```python
reg.fit(houseData[['Area', 'Bedroom', 'Age']], houseData.Price)
```

#### Slope

```python
reg.coef_
```

```python
reg.predict([[1900]])
```

#### Intercept

```python
reg.intercept_
```

#### Predict the Sales Price Of the House

**3000 sqr ft area, 3 bedrooms, 50 years old**

```python
reg.predict([[3000, 3, 50]])
```

#### Manually checking

$
Y = a + b_1 X_1 + b_2 X_2 + … + b_kX_k
$

```python
y = reg.coef_[0] * 3000 + reg.coef_[1] * 3 + reg.coef_[2] * 50 + reg.intercept_
y
```

#### Predict the Sales Price Of the House

**2500 sqr ft area, 4 bedrooms, 21 years old**

```python
reg.predict([[2500, 4, 21]])
```
