### Tokenize

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
total_words = len(tokenizer.word_index) + 1
```

### Create input sequence

```python
input_sequences = []
for line in corpus:
  token_list = tokenizer.texts_to_sequences([line])[0]
  for i in range(1, len(token_list)):
    n_gram_sequence = token_list[:i+1]
    input_sequences.append(n_gram_sequence)
```

### Pad sequences

```python
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))
print(input_sequences)
```

### Split features and labels

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
y = to_categorical(y, num_classes=total_words)
```

```python
model = Sequential([
    Embedding(total_words, 10, input_length=max_sequence_len - 1),
    SimpleRNN(100),
    Dense(total_words, activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X, y, epochs=500, verbose=1)
```

```python
def predict_next_word(seed_text):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')
    predicted_prob = model.predict(token_list, verbose=0)[0]
    predicted_index = np.argmax(predicted_prob)
    for word, index in tokenizer.word_index.items():
        if index == predicted_index:
            return word
    return ""
```

```python
def predict_next_word2(prefix):
    prefix = prefix.strip().lower()
    prefix_tokens = tokenizer.texts_to_sequences([prefix])[0]

    completions = set()
    for line in corpus:
        line_lower = line.lower()
        tokens = tokenizer.texts_to_sequences([line_lower])[0]
        # Check if line starts with prefix tokens
        if tokens[:len(prefix_tokens)] == prefix_tokens:
            # Get the remaining words after the prefix
            remaining_tokens = tokens[len(prefix_tokens):]
            if remaining_tokens:
                # Convert tokens back to words
                words_after_prefix = []
                for idx in remaining_tokens:
                    for word, index in tokenizer.word_index.items():
                        if index == idx:
                            words_after_prefix.append(word)
                            break
                # Join them with space and add '...' before it
                completion = '... ' + ' '.join(words_after_prefix)
                completions.add(completion)

    return list(completions)

print("Enter your phrase (type 'exit' to quit):")

while True:
    user_input = input("> ").strip()
    if user_input.lower() == 'exit':
        print("Goodbye!")
        break

    results = predict_next_word2(user_input)
    if not results:
        print("No suggestions found.")
        continue

    print("Suggestions:")
    for r in results:
        print("  " + r)
```
