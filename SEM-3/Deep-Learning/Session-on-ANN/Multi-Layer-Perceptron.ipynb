## Importing Requiredd Libraries


```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
import pandas as pd
from google.colab import files
import warnings
warnings.filterwarnings('ignore')
```

### As this is regression problem, we are separating the independant and dependant variable

```python
X = diabetes_pd.iloc[:, 0:8]
y = diabetes_pd.iloc[:, 8]
```

```python
X
```

```python
y
```

### Sequential Model Building by Adding Layers in NN

```python
model = Sequential()
model.add(Dense(8, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### Compiling the Model

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Fit the Network

```python
model.fit(X, y, epochs=1500, batch_size=10)
```

### Activation Function

| Task Type           | Activation Function      | Output Layer Configuration                     | Example Use Case           |
|---------------------|--------------------------|------------------------------------------------|---------------------------|
| Regression          | Linear                   | Number of neurons = number of outputs (usually 1) | Predicting house prices    |
| Binary Classification| Sigmoid                  | Number of neurons = 1                           | Spam detection (spam or not)|
| Multiclass Classification | Softmax              | Number of neurons = number of classes          | Handwritten digit recognition (0-9), Iris flower Classification |


### Standard Loss Function

| Task Type             | Loss Function         | Description                                         | Example Use Case               |
|-----------------------|-----------------------|-----------------------------------------------------|-------------------------------|
| Regression            | Mean Squared Error (MSE) | Measures average squared difference between predicted and actual values | Predicting house prices        |
| Binary Classification | Binary Cross-Entropy  | Measures the difference between predicted probabilities and actual binary labels | Spam detection (spam or not)  |
| Multiclass Classification | Categorical Cross-Entropy | Measures the difference between predicted class probabilities and actual class labels | Handwritten digit recognition (0-9), Classifying types of flowers |

### Evaluate the Network

```python
loss, accuracy = model.evaluate(X, y)
print(f"Loss: {loss}, Accuracy: {accuracy*100}")
```

### Make Predictions

```python
probabilities = model.predict(X)
```

```python
print(f"\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy*100))
```

## Assignment (Regression)

**Use Case:** Predicting continuous values, such as house prices, temperature, etc.

**Activation Function:** Linear activation function (linear)

**Output Layer:** Number of neurons matching the number of outputs (usually 1 for single-output regression)

**Example:** Predicting house prices based on features like size, number of bedrooms, and location.

```python
X2 = np.array([[1500,3],[1700,4],[1200,2],[1800,3],[1600,3]])
#target price
y2 = np.array([300000, 400000, 500000, 600000, 700000])
```

```python
regression_model = Sequential()
regression_model.add(Dense(1, input_dim=2, activation='linear'))
```

```python
regression_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
regression_model.fit(X2, y2, epochs=500, batch_size=1, verbose=0)
```

```python
# Predict prices for the input data
predictions = regression_model.predict(X2)
print("Predicted prices:", predictions.flatten())
```

```python
# Evaluate the model
loss = regression_model.evaluate(X2, y2, verbose=0)
print(f"Loss (MSE): {loss}")
```
