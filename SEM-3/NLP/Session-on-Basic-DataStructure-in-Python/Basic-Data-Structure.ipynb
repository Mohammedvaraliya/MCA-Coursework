### All type of Functions in Python

```python
# Built-in Functions
print(len("Hello"))
max(10, 20)
```

```python
# User-defined Functions
def greet():
    print("Hello!")
greet()
```

```python
#  Function with Parameters
def add(a, b):
    return a + b

print(add(2, 3))
```

```python
# Function with Default Parameters
def greet(name="User"):
    print(f"Hello, {name}!")

greet()
greet("Mohammed Varaliya")
```

```python
# Function with Return Value
def square(x):
    return x * x

print(square(4))
```

```python
# Lambda Functions (Anonymous Functions)
square = lambda x: x * x
print(square(5))
```

```python
# Recursive Functions
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
# Nested Functions (Inner Functions)
def outer():
    def inner():
        print("Inner function")
    inner()

outer()
```

```python
# Higher-order Functions
def shout(text):
    return text.upper()

def greet(func):
    message = func("hello")
    print(message)

greet(shout)
```

```python
# Generator Functions
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)
```

```python
# Decorator Functions
def decorator(func):
    def wrapper():
        print("⭐⭐⭐⭐⭐⭐")
        func()
        print("⭐⭐⭐⭐⭐⭐")
    return wrapper

@decorator
def say_hi():
    print("Hi")

say_hi()
```

```python
import time

# Decorator Functions
def decorative_func(func):
    def wrapper(*args, **kwargs):
      start = time.time()
      result = func(*args, **kwargs)
      end = time.time()
      print(func.__name__, "took", {end - start}, "miliseconds")
      return result
    return wrapper

@decorative_func
def add(*args):
    return sum(args)

@decorative_func
def wish(**kwargs):
    return f"wish: ", kwargs

@decorative_func
def doSomething(*nums):
  for num in nums:
    num = num + 2
  return nums
```

```python
wish(name="KJ Somaiya Institute of Management", city="Mumbai Metropolitan city")
doSomething(9, 1)
```

```python
# Asynchronous Functions
import asyncio

async def async_func(*args):
  for v in args:
    print(f"Value: {v}")
    await asyncio.sleep(2)

await async_func(1, 2, 3, 4)
```

```python
# Asynchronous Functions
import asyncio

async def async_func2(**kwargs):
  for v in kwargs:
    print(f"Value: {kwargs[v]}")
    await asyncio.sleep(2)

await asyncio.gather(async_func(1, 2, 3, 4), async_func2(name="Mohammed Varaliya", city="Mumbai"))
```
