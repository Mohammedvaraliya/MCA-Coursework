## Create a dataset using Faker and custom-made fields with random choice

1. Name  
1. Address  
1. Country  
1. City (random from custom list)  
1. Gender (random from custom list)  
1. Course (random from custom list)

```python
city_data = ["Mumbai", "Delhi", "Bangalore", "Hyderabad", "Chennai", "Pune"]
gender_data = ["Male", "Female", "Non-binary", "Other"]
course_data = ["Computer Science", "Business", "Physics", "Mathematics", "Economics", "MCA", "History"]
```

```python
data = []

for _ in range(100):
    row = {
        "Name": faker.name(),
        "Address": faker.address().replace("\n", ", "),
        "Country": faker.country(),
        "City": random.choice(city_data),
        "Gender": random.choice(gender_data),
        "Course": random.choice(course_data)
    }
    data.append(row)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv("synthetic_student_data.csv", index=False)
```

## E-Commerce Dataset

```python
faker.add_provider(faker_commerce.Provider)
```

```python
positive_words = [
    "excellent", "amazing", "fantastic", "great", "love", "wonderful",
    "perfect", "high quality", "satisfied", "happy", "fast delivery",
    "recommend", "awesome", "best", "friendly service"
]

negative_words = [
    "bad", "poor", "terrible", "disappointed", "slow", "not recommend",
    "worst", "broken", "low quality", "unhappy", "rude service",
    "defective", "late delivery", "problem", "frustrated"
]

positive_phrases = [
    "Exceeded my expectations",
    "Would buy again",
    "Highly recommend this product",
    "Customer service was excellent",
    "Quality is outstanding",
    "Very satisfied with the purchase",
    "Fast shipping and well packaged",
    "Product works perfectly",
    "Great value for money",
    "Loved the design and features"
]

negative_phrases = [
    "Did not meet expectations",
    "Would not buy again",
    "Customer service was unhelpful",
    "Product arrived damaged",
    "Poor packaging",
    "Took too long to arrive",
    "Stopped working after a week",
    "Not worth the price",
    "Very disappointed with the quality",
    "Had many issues with this item"
]
```

```python
def generate_review_with_product(type, product):
    if type == "positive":
        words = random.sample(positive_words, 10)
        phrase = random.choice(positive_phrases)
        review = f"{phrase} for the product '{product}'. Overall, it is {' '.join(words)}."
        return review
    else:
        words = random.sample(negative_words, 10)
        phrase = random.choice(negative_phrases)
        review = f"{phrase} for the product '{product}'. Unfortunately, it is {' '.join(words)}."
        return review
```

```python
data = []

for _ in range(50):
    sentiment = random.choice(["positive", "negative"])
    product_name = faker.ecommerce_name()
    review_text = generate_review_with_product(sentiment, product_name)
    row = {
        "Product Name": product_name,
        "Review": review_text,
        "Sentiment": sentiment
    }
    data.append(row)
```

```python
ecommerce_df = pd.DataFrame(data)
```

```python
ecommerce_df
```

```python
ecommerce_df.to_csv("ecommerce_dataset.csv", index=False)
```
