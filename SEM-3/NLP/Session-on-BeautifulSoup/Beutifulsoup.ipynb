```python
def scrape_page(page_number):
    url = f'https://quotes.toscrape.com/page/{page_number}/'
    response = requests.get(url)
    if response.status_code != 200:
        return None

    html_content = BeautifulSoup(response.content, 'html.parser')

    all_spans = html_content.select("span.text")
    all_author_spans = html_content.select("small.author")
    all_div_tags = html_content.select("div.tags")

    all_quotes = [span.text for span in all_spans]
    all_authors = [span.text for span in all_author_spans]
    all_tags = [", ".join([a.get_text() for a in div.select("a.tag")]) for div in all_div_tags]

    return {
        "quotes": all_quotes,
        "authors": all_authors,
        "tags": all_tags
    }

def scrape_pages(selected_pages):
    all_quotes = []
    all_authors = []
    all_tags = []

    for page in selected_pages:
        page_data = scrape_page(page)
        if page_data:
            all_quotes.extend(page_data["quotes"])
            all_authors.extend(page_data["authors"])
            all_tags.extend(page_data["tags"])

    quotes_dataset = pd.DataFrame({
        "Quotes": all_quotes,
        "Authors": all_authors,
        "Tags": all_tags
    })

    return quotes_dataset

page_selector = widgets.SelectMultiple(
    options=[f"Page {i}" for i in range(1, 11)],
    value=["Page 1"],
    description="Pages",
    disabled=False
)

def on_button_click(b):
    clear_output(wait=True)

    selected_pages = [int(page.split()[-1]) for page in page_selector.value]

    quotes_dataset = scrape_pages(selected_pages)

    display(HTML(quotes_dataset.to_html(index=False)))

    display(page_selector, button)

button = widgets.Button(description="Scrape Selected Pages")
button.on_click(on_button_click)

display(page_selector, button)
```
