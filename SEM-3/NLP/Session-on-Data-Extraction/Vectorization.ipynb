```python
tfid_vectorizer = TfidfVectorizer()
tfid_vectorizer
```

```python
matrix = tfid_vectorizer.fit_transform(english_reviews)
```

```python
# Convert matrix to 1D array
tfid_matrix = matrix.sum(0).A1
tfid_matrix
```

```python
words = tfid_vectorizer.get_feature_names_out()
words
```

```python
# Find out top 5 words used in the reviews
print(tfid_matrix.argsort()[::-1][:5])

for i in tfid_matrix.argsort()[::-1][:5]:
    print(words[i])
```

```python
# IDF values (optional)-
idf_matrix = tfid_vectorizer.idf_
idf_top_indices = idf_matrix.argsort()[::-1][:5]

print("\nTop 5 words with highest IDF values:\n")
for i in idf_top_indices:
    print(f"{words[i]}: {idf_matrix[i]:.4f}")
```

## Bigram, N-gram

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(2,2), stop_words="english")
bigrams = vectorizer.fit_transform(english_reviews)
print("bigrams", vectorizer.get_feature_names_out())
```

```python
vectorizer = CountVectorizer(ngram_range=(2,2), stop_words="english")
bigrams = vectorizer.fit_transform(english_reviews)

my_biagrams = vectorizer.get_feature_names_out()
my_biagrams

frequency = bigrams.sum(0).A1
frequency

for i in frequency.argsort()[::-1][:5]:
    print(f"biagram: {my_biagrams[i]} frequency: {frequency[i]}")
```

## Part of Speech Tagging

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from nltk import pos_tag

for i,doc in enumerate(english_reviews,1):
     print(f"review {i}:")
     # tokens = word_tokenize(doc)
     # pos_tags = pos_tag(tokens)
     tokens = word_tokenize(doc)
     pos_tags = pos_tag(tokens)
     print(pos_tags)
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")

for i,doc in enumerate(english_reviews,1):
    print(f"review {i}:")
    doc = nlp(doc)
    for token in doc:
        print(token.text, token.pos_)
    print("\n")
```
