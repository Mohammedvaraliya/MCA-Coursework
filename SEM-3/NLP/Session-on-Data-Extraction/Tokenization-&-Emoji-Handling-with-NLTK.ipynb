### Words Tokenization

```python
words = word_tokenize(text)
print("Word Tokens:\n", words[:50])
```

### Sentence Tokenization

```python
sentences = sent_tokenize(text)

cleaned_sentences = []

for sentence in sentences:
    # Remove URLs
    sentence = re.sub(r'http\S+', '', sentence)
    # Remove HTML tags
    sentence = re.sub(r'<.*?>', '', sentence)
    # Remove single quotes
    sentence = re.sub(r"'", '', sentence)
    # Remove punctuation
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    # strip extra whitespace
    cleaned_sentences.append(sentence.strip())

    print(sentence)
```

### Character Tokenization

```python
chars = list(text)
print("First 100 Characters:\n", chars[:100])
```

### Tokenization using TweetTokenizer (better for emojis and social media text)

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = []

for sent in sentences:
    tweet_tokens.extend(tweet_tokenizer.tokenize(sent))

tweet_tokens
```

### Converting emojis to text descriptions

```python
updated_text = ' '.join(cleaned_sentences)
```

```python
text_with_emoji_names = emoji.demojize(updated_text)
text_with_emoji_names
```
